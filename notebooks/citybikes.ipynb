# Cleaning the Data

```python
import pandas as pd
```

```python
citibike_data_1 = pd.read_csv(r'/Users/shashankhmg/Documents/AXA-Casestudy/Data-Science-Challenge/data/raw/citybike/202501-citibike-tripdata/202501-citibike-tripdata_1.csv')
citibike_data_2 = pd.read_csv(r'/Users/shashankhmg/Documents/AXA-Casestudy/Data-Science-Challenge/data/raw/citybike/202501-citibike-tripdata/202501-citibike-tripdata_2.csv')
citibike_data_3 = pd.read_csv(r'/Users/shashankhmg/Documents/AXA-Casestudy/Data-Science-Challenge/data/raw/citybike/202501-citibike-tripdata/202501-citibike-tripdata_3.csv')
```

```python
if list(citibike_data_1.columns) == list(citibike_data_2.columns) == list(citibike_data_3.columns):
    print('All 3 CSVs have the same data schema')
else:
    print('Column mismatch detected!')
```

```python
#concat all 3 df into one dataset 
# Combine all three DataFrames
citibike_combined = pd.concat([citibike_data_1, citibike_data_2, citibike_data_3], ignore_index=True)

# Display shape of the combined dataset
print("Combined DataFrame shape:", citibike_combined.shape)
```

```python
citibike_combined
```

```python
print("Duplicate rows:", citibike_combined.duplicated().sum())
```

```python
citibike_combined['started_at'] = pd.to_datetime(citibike_combined['started_at'])
citibike_combined['ended_at'] = pd.to_datetime(citibike_combined['ended_at'])
```

```python
#checking null values
citibike_combined.isnull().sum()
```

```python
# Drop rows where start & end station ID are missing (critical for analysis)
citibike_combined_cleaned = citibike_combined.dropna(subset=['start_station_id', 'end_station_id']).copy()

# Fill missing station names with "Unknown" - assuming the rider might have parked it on the road. 
citibike_combined_cleaned.loc[:, 'start_station_name'] = citibike_combined_cleaned['start_station_name'].fillna("Unknown")
citibike_combined_cleaned.loc[:, 'end_station_name'] = citibike_combined_cleaned['end_station_name'].fillna("Unknown")

# Fill missing end latitude & longitude with start latitude & longitude - because the rider might have ended the ride by mistake. 
citibike_combined_cleaned.loc[:, 'end_lat'] = citibike_combined_cleaned['end_lat'].fillna(citibike_combined_cleaned['start_lat'])
citibike_combined_cleaned.loc[:, 'end_lng'] = citibike_combined_cleaned['end_lng'].fillna(citibike_combined_cleaned['start_lng'])

# Check missing values after cleaning
print("Missing values after cleaning:", citibike_combined_cleaned.isnull().sum())
```

```python
# Save as CSV
citibike_combined_cleaned.to_csv(r"/Users/shashankhmg/Documents/AXA-Casestudy/Data-Science-Challenge/data/processed/cleaned_citibike_data.csv", index=False)
```

# EDA

```python
# Overview of the dataset
print("Dataset shape:", citibike_combined_cleaned.shape)
print("Columns:", citibike_combined_cleaned.columns)
```

```python
# Time-based Analysis (Peak Usage Trends)

#Rides per Day

import matplotlib.pyplot as plt

# Extract date from datetime column
citibike_combined_cleaned['date'] = citibike_combined_cleaned['started_at'].dt.date

# Group by date
daily_rides = citibike_combined_cleaned.groupby('date').size()

# Plot daily rides
plt.figure(figsize=(12, 5))
plt.plot(daily_rides, marker='o', linestyle='-')
plt.title("Daily CitiBike Usage")
plt.xlabel("Date")
plt.ylabel("Number of Rides")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
import seaborn as sns

# Extract day of the week (0=Monday, 6=Sunday)
citibike_combined_cleaned['weekday'] = citibike_combined_cleaned['started_at'].dt.dayofweek

# Map days to labels
day_labels = {0: "Monday", 1: "Tuesday", 2: "Wednesday", 3: "Thursday", 
              4: "Friday", 5: "Saturday", 6: "Sunday"}

citibike_combined_cleaned['weekday_name'] = citibike_combined_cleaned['weekday'].map(day_labels)

# Group by day of the week
weekday_rides = citibike_combined_cleaned.groupby('weekday_name').size()

# Sort to ensure correct order (Mon-Sun)
weekday_rides = weekday_rides.reindex(["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(x=weekday_rides.index, y=weekday_rides.values, palette="Blues_r")
plt.title("CitiBike Usage by Weekday vs. Weekend")
plt.xlabel("Day of the Week")
plt.ylabel("Number of Rides")
plt.grid(axis="y")
plt.show()
```

```python
#Peak Hours Analysis
# Extract hour from start time
citibike_combined_cleaned['hour'] = citibike_combined_cleaned['started_at'].dt.hour

# Group by hour
hourly_rides = citibike_combined_cleaned.groupby('hour').size()

# Plot
plt.figure(figsize=(10, 5))
plt.bar(hourly_rides.index, hourly_rides.values)
plt.title("Rides per Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Rides")
plt.xticks(range(0, 24))
plt.grid()
plt.show()
```

```python
# User Analysis (Casual vs. Member Riders)


# Count the number of rides for each user type
user_counts = citibike_combined_cleaned['member_casual'].value_counts()

# Plot user distribution
plt.figure(figsize=(7, 5))
sns.barplot(x=user_counts.index, y=user_counts.values, palette="Blues_r")

# Labels and title
plt.title("CitiBike Users: Subscribers vs. Casual Riders", fontsize=14)
plt.xlabel("User Type", fontsize=12)
plt.ylabel("Number of Rides", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis="y")

# Display the plot
plt.show()
```

```python
# Calculate trip duration in minutes

citibike_combined_cleaned['trip_duration'] = (citibike_combined_cleaned['ended_at'] - citibike_combined_cleaned['started_at']).dt.total_seconds() / 60

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(citibike_combined_cleaned['trip_duration'], bins=50, range=[0, 60], alpha=0.7, color='blue')
plt.title("Trip Duration Distribution")
plt.xlabel("Minutes")
plt.ylabel("Number of Rides")
plt.grid()
plt.show()
```

```python
# Most Popular Start Stations
top_stations = citibike_combined_cleaned['start_station_name'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_stations.values, y=top_stations.index, palette="Blues_r")
plt.title("Top 10 Most Popular Start Stations")
plt.xlabel("Number of Rides")
plt.ylabel("Station Name")
plt.show()
```

```python
# Geospatial Analysis (Mapping Stations & Routes)
import geopandas as gpd

# Define NYC boundary (approximate)
NYC_LAT_MIN, NYC_LAT_MAX = 40.5, 40.9
NYC_LNG_MIN, NYC_LNG_MAX = -74.2, -73.7

# Filter out incorrect coordinates
valid_data = citibike_combined_cleaned[
    (citibike_combined_cleaned['start_lat'].between(NYC_LAT_MIN, NYC_LAT_MAX)) &
    (citibike_combined_cleaned['start_lng'].between(NYC_LNG_MIN, NYC_LNG_MAX))
].copy()

# Create a GeoDataFrame
gdf = gpd.GeoDataFrame(valid_data, geometry=gpd.points_from_xy(valid_data['start_lng'], valid_data['start_lat']))

# Plot station locations
fig, ax = plt.subplots(figsize=(10, 6))
gdf.plot(ax=ax, alpha=0.5, marker='o', color='blue')

plt.title("CitiBike Start Locations (Filtered for NYC)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid()
plt.show()
```

```python
# Overlay CitiBike Start & End Locations on a Map

# Define NYC boundary for valid coordinates
NYC_LAT_MIN, NYC_LAT_MAX = 40.5, 40.9
NYC_LNG_MIN, NYC_LNG_MAX = -74.2, -73.7

# Filter valid start locations
valid_start = citibike_combined_cleaned[
    (citibike_combined_cleaned['start_lat'].between(NYC_LAT_MIN, NYC_LAT_MAX)) &
    (citibike_combined_cleaned['start_lng'].between(NYC_LNG_MIN, NYC_LNG_MAX))
].copy()

# Filter valid end locations
valid_end = citibike_combined_cleaned[
    (citibike_combined_cleaned['end_lat'].between(NYC_LAT_MIN, NYC_LAT_MAX)) &
    (citibike_combined_cleaned['end_lng'].between(NYC_LNG_MIN, NYC_LNG_MAX))
].copy()

# Create GeoDataFrames for start and end locations
gdf_start = gpd.GeoDataFrame(valid_start, geometry=gpd.points_from_xy(valid_start['start_lng'], valid_start['start_lat']))
gdf_end = gpd.GeoDataFrame(valid_end, geometry=gpd.points_from_xy(valid_end['end_lng'], valid_end['end_lat']))

# Plot start and end locations
fig, ax = plt.subplots(figsize=(10, 6))
gdf_start.plot(ax=ax, alpha=0.5, marker='o', color='blue', label="Start Locations")
gdf_end.plot(ax=ax, alpha=0.5, marker='x', color='red', label="End Locations")

# Add title and labels
plt.title("CitiBike Start vs. End Locations (NYC)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend()
plt.grid()
plt.show()
```
